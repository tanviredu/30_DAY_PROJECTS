```python
file_path = "/root/.cache/kagglehub/datasets/parasharmanas/movie-recommendation-system/versions/1" + "/" +"movies.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df['title'] = df['title'].str[:-6].str.strip()
```

```python
df.head()
```

```python
df = df.set_index("movieId")
```

```python
df = df[:2500]
```

```python
tfIdf = TfidfVectorizer(stop_words="english")
```

```python
tf_Idf_matrix = tfIdf.fit_transform(df["genres"])
```

```python
cosine_sim = cosine_similarity(tf_Idf_matrix, tf_Idf_matrix)
```

```python
def get_recommendation(title,cosine_sim = cosine_sim):
  #get the index of the movie
  idx = df[df["title"]== title].index[0]
  print(idx)
  sim_score = list(enumerate(cosine_sim[idx]))
  sim_score = sorted(sim_score, key=lambda x: x[1], reverse=True)
  sim_score = sim_score[1:10]
  movie_indices = [i[0] for i in sim_score]
  return df[['title', 'genres']].iloc[movie_indices]
```

```python
movie_title = "Toy Story"
recommended_movie = get_recommendation(movie_title)
print(recommended_movie)
```
